```python
import requests
import lxml
from bs4 import BeautifulSoup as bs

url = 'http://www.kma.go.kr/weather/forecast/mid-term-rss3.jsp?stnId=108'
req = requests.get( url )

print( req.status_code )

soup = bs( req.content, 'lxml' )

# print( soup.prettify()  )

content = ''

for location in soup.select('location'):
    content += '<h3>{}</h3>'.format( location.select_one( 'city' ).text )
    content += '날씨 : {}<br />'.format( location.select_one( 'wf' ).text )
    content += '최저/최고 기온 : {}/{}'.format( 
                                        location.select_one( 'tmn' ),
                                        location.select_one( 'tmx' ) 
                                        )
    content += '<hr />'
    
with open( 'weather.html', 'w' ) as f:
    f.write( content )
```

```python
def test( function ):
    def wrapper():
        print( "Start Wrapper" )
        function()
        print( "End Wrapper" )
    return wrapper

@test
def greeting():
    print( "hello" )
    
greeting()
```

```python
class Circle:
    def __init__(self, radius):
        self.__radius = radius
        
    @property
    def radius(self):
        return self.radius
    
    @radius.setter
    def radius(self, value):
        if( value < 0 ):
            raise TypeError( "Not allowed negative number" )
        self.__radius = value
```
